```python
import pandas as pd
import matplotlib.pyplot as plt

# The 'Close' column was dropped from 'df'. Using 'data1' which was loaded before the drop.
# However, 'data1' has messy headers due to 'to_csv' and 'read_csv' handling of multi-index.
# The actual data starts from row index 1, and the 'Date' and ' Close' (with a space) columns hold the data.

# Create a temporary DataFrame for plotting from data1, starting from the actual data rows
plot_df = data1.loc[1:].copy()

# Ensure 'Date' column is in datetime format and ' Close' column is numeric
plot_df['Date'] = pd.to_datetime(plot_df['Date'])
plot_df[' Close'] = pd.to_numeric(plot_df[' Close'])

plt.figure(figsize=(12,6))
plt.plot(plot_df['Date'], plot_df[' Close'], label=f'{stock}Closing Price', linewidth=2)
plt.title(f'{stock}Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.show()
```
